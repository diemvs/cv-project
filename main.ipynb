```python
import cv2
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

def compare_yolo_models(image_dir, model1_path, model2_path, conf=0.5, save_results=False):
    """
    Прогоняет все изображения из директории через две модели YOLO и выводит их на графике для сравнения.
    
    :param image_dir: Путь к папке с изображениями
    :param model1_path: Путь к первой модели YOLO (например, yolov8n.pt)
    :param model2_path: Путь ко второй модели YOLO (например, best.pt)
    :param conf: Минимальный порог уверенности для предсказаний
    :param save_results: Если True, сохраняет результаты в папку 'results/'
    """
    # Загружаем обе модели
    model1 = YOLO(model1_path)
    model2 = YOLO(model2_path)

    # Создаем папку для сохранения результатов, если нужно
    if save_results:
        os.makedirs("results", exist_ok=True)

    # Получаем список изображений в папке
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
    
    if not image_files:
        print("В папке нет изображений!")
        return

    # Количество изображений для обработки
    num_images = len(image_files)
    fig, axes = plt.subplots(num_images, 2, figsize=(10, 5 * num_images))  # Две колонки: одна для каждой модели

    if num_images == 1:
        axes = [axes]  # Делаем axes списком, если одно изображение

    # Обрабатываем каждое изображение
    for idx, img_name in enumerate(image_files):
        img_path = os.path.join(image_dir, img_name)

        # Загружаем изображение
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Прогоняем через обе модели
        results1 = model1.predict(img_path, conf=conf)
        results2 = model2.predict(img_path, conf=conf)

        # Функция для отрисовки боксов
        def draw_boxes(image, results, color=(255, 0, 0)):
            for result in results:
                for box in result.boxes:
                    x1, y1, x2, y2 = map(int, box.xyxy[0])
                    conf_score = box.conf[0].item()
                    cls = int(box.cls[0].item())
                    label = f"{model1.names[cls]}: {conf_score:.2f}"

                    cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
                    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
            return image

        # Копируем изображения для отрисовки
        img1 = img.copy()
        img2 = img.copy()

        # Рисуем боксы
        img1 = draw_boxes(img1, results1, color=(255, 0, 0))  # Синяя рамка для первой модели
        img2 = draw_boxes(img2, results2, color=(0, 255, 0))  # Зеленая рамка для второй модели

        # Отображаем изображения
        axes[idx][0].imshow(img1)
        axes[idx][0].set_title(f"Model 1: {model1_path}")
        axes[idx][0].axis("off")

        axes[idx][1].imshow(img2)
        axes[idx][1].set_title(f"Model 2: {model2_path}")
        axes[idx][1].axis("off")

        # Сохраняем результаты, если нужно
        if save_results:
            cv2.imwrite(f"results/{img_name}_model1.jpg", cv2.cvtColor(img1, cv2.COLOR_RGB2BGR))
            cv2.imwrite(f"results/{img_name}_model2.jpg", cv2.cvtColor(img2, cv2.COLOR_RGB2BGR))

    plt.tight_layout()
    plt.show()
```

```python
compare_yolo_models(
    image_dir="test_images", 
    model1_path="yolov8n.pt", 
    model2_path="runs/train_2/weights/best.pt",  
    conf=0.3,
    save_results=False 
)
```
